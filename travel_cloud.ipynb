```python
#開發一個簡單的爬蟲程式，爬取「ETtoday旅遊雲」網頁，擷取旅遊景點的標題資訊
import requests
import pprint as pp
from bs4 import BeautifulSoup

response = requests.get("https://travel.ettoday.net/category/%E6%96%B0%E5%8C%97/")

soup= BeautifulSoup(response.text,'lxml')
print(f'*列印標題title且不含標籤: {soup.title.text}*')

#直接列印<h3 itemprop="headline">結果
pp.pprint(soup.find_all("h3", itemprop="headline"))

print("----分隔線----")

#使用getText()方式取得結果:沒有子標籤

titles= soup.find_all("h3", itemprop="headline")
for data in titles:
    print(data.getText())

print("----分隔線----")

print(soup.prettify())

```

```python
#find title and paragraph of article, limit 3 cases
#compare output result by using pprint
#同時搜尋多個HTML標籤，可以將標籤名稱打包成串列(List)後，傳入find_all()方法

pp.pprint(soup.find_all(["h3","p"],limit=3))

print("----分隔線----")
result= soup.find_all(["h3","p"],limit=3)
print(result)
```

```python
#如果某一節點下有多個子節點時，則使用select()方法(Method)，選取子節點
#利用select()方法(Method)，選取<div>其下所有的<a>標籤，並且為串列(List)的資料型態。

result1 = soup.find_all("div", itemprop="itemListElement", limit=3)
for res in result1:
    print(res.select("a"))

print("----分隔線----")
#選取<div>下符合<a class="pic">的標籤

pp.pprint(soup.select("div a.pic"))

print("----分隔線----")

#use loop 訪問<div><a class="pic>其下各個 href
result2=[]
for a in soup.select("div a.pic"):
    result2.append(a.get('href')) #a.get("") or a[""]
pp.pprint(result2)
```
